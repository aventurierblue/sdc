```python
import pandas as pd
import cupy as cp
import numpy as np
```

```python
allen_table = pd.read_csv("newsdc_cahn_table.csv", index_col=0)
```

```python
allen_table
```

```python
for i in np.arange(0, len(allen_table.index)):
    row = allen_table.iloc[i]
    nvars = row["nvars"]
    dt = row["dt"]
    quadn = row["quad_nodes"]
    new_max = row["sdc_max"]
    sdc_sum = row["sdc_sum"]
    new_sum = row["new_sum"]
    norm = row["norm"]
    print(f"{quadn} & {nvars} & {new_max} & {sdc_sum} & {new_sum} & {norm:.1e}\\\\")
```
